## 引入相关数据分析包
* pandas，用于数据集读取，处理和分析
* numpy，用于数据集处理
* matplotlib，用于数据可视化

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('./dataset/NationalNames.csv')
```

```python
def plotname(name, gender):
    """
        对姓名及性别进行可视化
    """
    # 找到符合条件的数据
    data_named = data[(data.Name==name) & (data.Gender==gender)]
        
    plt.figure(figsize=(10,6))    
    # 对数据的年份，数量进行可视化
    plt.plot(data_named.Year,data_named.Count,'g-')
    plt.title('%s name variation over time'%name)
    plt.ylabel('counts')
    plt.xticks(data_named.Year,rotation='vertical')
    
    # 设置x轴数值范围
    plt.xlim([1985,2015])    
    plt.show() 
```

```python
# 老友记 姓名影响分析
friends_characters = [('Rachel','F'),('Monica','M'),('Phoebe','F'),
                      ('Ross','F'),('Chandler','F'),('Joey','F')]
```

```python
for character in friends_characters:
    plotname(character[0],character[1])
```

```python
# 冰火歌 和 生活大爆炸 姓名影响分析
more_characters = [('Khaleesi', 'F'), ('Sheldon', 'M')]
```

```python
for character in more_characters:
    plotname(character[0],character[1])
```
